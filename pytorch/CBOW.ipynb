```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

```

```python
CONTEXT_SIZE = 2  # 2 words to the left, 2 to the right
raw_text = """We are about to study the idea of a computational process.
Computational processes are abstract beings that inhabit computers.
As they evolve, processes manipulate other abstract things called data.
The evolution of a process is directed by a pattern of rules
called a program. People create programs to direct processes. In effect,
we conjure the spirits of the computer with our spells.""".split()

# By deriving a set from `raw_text`, we deduplicate the array
vocab = set(raw_text)
vocab_size = len(vocab)

word_to_ix = {word: i for i, word in enumerate(vocab)}
data = []
for i in range(2, len(raw_text) - 2):
    context = [raw_text[i - 2], raw_text[i - 1],
               raw_text[i + 1], raw_text[i + 2]]
    target = raw_text[i]
    data.append((context, target))
print(data[:5])


class CBOW(nn.Module):

    def __init__(self,vocab_size,embedding_dim):
        super(CBOW, self).__init__()
        self.em=nn.Embedding(num_embeddings=vocab_size,embedding_dim=embedding_dim)
        self.fnn1=nn.Linear(embedding_dim*4,128)
        self.fnn2=nn.Linear(128,vocab_size)

    def forward(self, inputs):
        embeding=self.em(inputs).view(-1)
        layer1=F.relu(self.fnn1(embeding))
        layer2=self.fnn2(layer1)
        return layer2
    
    
# create your model and train.  here are some functions to help you make
# the data ready for use by your module
loss_function =nn.CrossEntropyLoss()
model= CBOW(vocab_size=vocab_size,embedding_dim=16)
optimizer = optim.SGD(model.parameters(), lr=0.001)
for i in range(300):
    for inputs,target in data:
        inputs_idx=[word_to_ix[w] for w in inputs]
        target=word_to_ix[target]
    
        inputs_idx=torch.tensor(inputs_idx, dtype=torch.long)
        target=torch.tensor(target, dtype=torch.long)
        
        model.zero_grad()
        logit=model(inputs_idx)

        loss=loss_function(logit.view((1,-1)),target.view((1)))
        loss.backward()
        
        optimizer.step()
        print('epcoh{} loss------------->{}'.format(i,loss))
```

```python
with torch.no_grad():
    We=model.em(torch.LongTensor([word_to_ix['We']]))
    computers=model.em(torch.LongTensor([word_to_ix['computers.']]))
    processes=model.em(torch.LongTensor([word_to_ix['processes']]))
    process=model.em(torch.LongTensor([word_to_ix['process']]))
    print(F.cosine_similarity(processes,computers),F.cosine_similarity(We,processes),F.cosine_similarity(process,processes))
```
